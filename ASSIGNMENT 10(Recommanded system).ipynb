# Importing libraries 

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('book.csv',encoding='latin-1')
df
```

# Convert the dataset into a matrix

```python
vectorizer = CountVectorizer()
book_matrix = vectorizer.fit_transform(df['Book.Title'])
print(book_matrix)
```

```python
# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(book_matrix, book_matrix)
```

```python
# Function to get book recommendations based on cosine similarity
def get_recommendations(book_title, cosine_sim_matrix, df):
    # Check if the book title exists in the dataframe
    if book_title in df['Book.Title'].values:
        book_index = df[df['Book.Title'] == book_title].index[0]
        sim_scores = list(enumerate(cosine_sim[book_index]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:]  # Exclude the book itself (index 0)

        book_indices = [x[0] for x in sim_scores]
        # Return the top 10 most similar books
        return df['Book.Title'].iloc[book_indices]
    else:
        return "The book.title you entered does not exist in the database. Please try again with a different title."

book_title = input("Enter a book.title: ")
# Example: Get recommendations for 'Cannibal In Manhattan'
recommendations = get_recommendations(book_title, cosine_sim, df)

print(f"Recommendations for {book_title}:")
print(recommendations)
```
